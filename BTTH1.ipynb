```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import randint
import warnings
warnings.filterwarnings('ignore')
```

### 1_Quá trình X(n) Ber(p=0.5)

```python
x=[]
y=[]
for i in range(100):
    n=randint(0,1)
    x.append(i)
    y.append(n)
plt.figure(figsize=(12,4))
sns.lineplot(
    x=x,
    y=y,
    color='green',
    linestyle='-'
)
plt.title('Qua trinh Ber(p=0.5)',fontsize=20,color='r')
plt.xlabel('n',fontsize=15)
plt.ylabel('X(n)',fontsize=15)
plt.xticks(range(0,110,10))
plt.yticks(range(0,2,1))
plt.grid(True)
plt.show()
```

### 2_Quá trình bước nhảy Dn với Ber(p=0.5)

```python
x=[]
y=[]
for i in range(100):
    n=randint(0,1)
    Dn=2*n-1
    x.append(i)
    y.append(Dn)
plt.figure(figsize=(12,4))
sns.lineplot(
    x=x,
    y=y,
    color='orange'
)
plt.title('Qua trinh Dn Ber(p=0.5)',fontsize=20,color='r',loc='left')
plt.xlabel('n',fontsize=15)
plt.ylabel('Dn',fontsize=15)
plt.xticks(range(0,110,10))
plt.yticks(range(-1,2,1))
plt.grid(True)
plt.show()
```

### 3_Quá trình nhị thức Y(n) với Ber(p=0.5)

```python
Yn=0
x=[]
y=[]
for i in range(50):
    n=randint(0,1)
    Yn+=n
    x.append(i)
    y.append(Yn)
plt.figure(figsize=(8,8))
sns.lineplot(
    x=x,
    y=y,
    color='green'
)
plt.title('Qua trinh Yn Ber(p=0.5)',fontsize=20,color='r',loc='left')
plt.xlabel('n',fontsize=15)
plt.ylabel('Yn',fontsize=15)
plt.xticks(range(0,51,10))
plt.grid(True)
plt.show()
plt.show()
```
